# Import Libraries 

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Define Dataset Path

```python
DATA_PATH = "./data"

categories = ["geometric", "real_world", "complex"]
```

# Image Loader Function

```python
def load_images(data_path, categories):
    images = {}
    
    for category in categories:
        folder_path = os.path.join(data_path, category)
        images[category] = []
        
        for file in os.listdir(folder_path):
            img_path = os.path.join(folder_path, file)
            img = cv2.imread(img_path)
            
            if img is not None:
                img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
                images[category].append(img)
    
    return images
```

# Load Dataset

```python
dataset = load_images(DATA_PATH, categories)

for category in dataset:
    print(f"{category}: {len(dataset[category])} images")
```

# Display Sample Images

```python
def show_sample_images(dataset, category, num=3):
    plt.figure(figsize=(12,4))
    
    for i in range(num):
        plt.subplot(1, num, i+1)
        plt.imshow(dataset[category][i])
        plt.axis("off")
    
    plt.suptitle(f"Sample Images from {category}")
    plt.show()

show_sample_images(dataset, "geometric")
```
